```python
# importing RFE library
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)

rfe = RFE(lm, n_features_to_select=15)
rfe = rfe.fit(x_train, y_train)
```

# Model 1

```python
# importing statsmodels to add constant/intercept.

import statsmodels.api as sm

# Add a constant
x_train_lm1 = sm.add_constant(x_train_rfe)

# Create a first fitted model
lr1 = sm.OLS(y_train, x_train_lm1).fit()

lr1.summary()
```

The cut-off for p-value is 0.05. 
From above result we can see that p-value of all variables is less than 0.05. 
Let's check VIF for all variables.

```python
# importing libraries to check the VIF values of the feature variables.

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

vif = pd.DataFrame()
vif['Features'] = x_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(x_train_rfe.values, i) for i in range(x_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

It is clear that VIF of humidity is highest of all features.

# Final Model Interpretation

**Hypothesis Testing:**

Hypothesis Testing states that
H0 : B1=B2=B3=....=Bn=0
H1 : at least one of Bi != 0

Model 7 Co-efficient Values:
1. const	            0.1009
2. temp	                0.4782
3. windspeed	       -0.1771
4. season_spring       -0.0509
5. season_summer	    0.0665
6. season_winter	    0.0884
7. weather_Good/Clear	0.0944
8. month_Sept	        0.0884
9. year_2019	        0.2378	
10. holiday_Yes	       -0.0906	

Inference: From the summary of model 7 we can see that none of the co-efficients are zero. Hence, we can reject the null hypothesis.

**F Statistics** : F-Statistics is used for testing the overall significance of the Model: Higher the F-Statistics, more significant the Model is.

- F-statistic: 237.5
- Prob (F-statistic): 3.04e-174

**The F-Statistics value of 237.5 (which is greater than 1) and the p-value of '~0.0000' states that the overall model is significant**

### The equation for best fitted line based on Model 7:


$Count = 0.1009 + (temp * 0.4782) - (windspeed * 0.1771) + -(season_spring*0.0509)+(season_summer * 0.0665) + (season_winter * 0.0884) + (year_2019 * 0.2378) + (month_Sept * 0.0884) - (holiday Yes * 0.0906) + (weather_Good/Clear * 0.0944)$

### Interpretation of Coefficients:

- **const**: The Constant value 0.1009 indicates that, in the absence of all other predictor variables (i.e. when x1,x2...xn =0), the bike demand will still increase by 0.0417 units.

- ****: A coefficient value of 0.4728 indicates that, a unit increase in temp variable, increases the bike demand by 0.5428 units.

### Adjusted R^2 Value for TEST

$Adj r2 = 1-(1-R2)*(n-1)/(n-p-1)$

```python
# n for test data ,n1 for train data is number of rows
n = x_test.shape[0]
n1 = x_train_rfe7.shape[0]

# Number of features (predictors, p for test data, p1 for train data) is the number of columns
p = x_test.shape[1]
p1 = x_train_rfe7.shape[1]


# We find the Adjusted R-squared using the formula

adjusted_r2_test = 1-(1-r2_test)*(n-1)/(n-p-1)
adjusted_r2_train = 1-(1-r2_train)*(n1-1)/(n1-p1-1)

print('Test data adjusted r^2 :',round((adjusted_r2_test*100),2))
print('Train data adjusted r^2 :',round((adjusted_r2_train*100),2))
```
